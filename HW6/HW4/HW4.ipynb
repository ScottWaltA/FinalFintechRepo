```python
#Importing the libraries

import pandas as pd
import numpy as np
import datetime as dt
import seaborn as sns
from pathlib import Path
%matplotlib inline
```

```python
#All the operations on the whale_returns and plot the graphs

whale_path = Path('whale_returns2.csv')
whale_returns = pd.read_csv(whale_path, index_col="Date",infer_datetime_format=True,parse_dates=True)
whale_returns.dropna()
whale_returns.isnull().mean() * 100
whale_returnspct= whale_returns.pct_change()
whale_returnspct.plot(title="Whale Returns Percent Change")
whale_returns.plot(title="Whale Returns Total Returns")
whale_returnsbox = whale_returns.boxplot()
whale_returnsbox.plot(title="Total returns boxplot")
whale_returns = whale_returns.rolling(window=180).sum()
whale_returns.plot(title="Whale Rolling Returns")
whale_standard_deviation = whale_returns.rolling(window=180).std()
whale_standard_deviation.plot(title="Whale Returns standard Deviation")
whale_returnsewa = whale_returns.ewm(span=4, adjust=False).mean()
whale_returnsewa.plot(title="exponentially weighted movinng averages")
whale_returns.to_csv('whale_returnscsv.csv')
whale_returns = whale_returns.sort_index()
```

```python
whale_returns=whale_returns.dropna().copy()
```

```python
#all the operations on the algo_returns and plotting the graphs


algo_path = Path('algo_returns.csv')
algo_returns = pd.read_csv(algo_path, index_col="Date",infer_datetime_format=True,parse_dates=True)
algo_returnspct=algo_returns.pct_change()
algo_returnspct.plot(title="Algo Returns Percent Change")
algo_returns.plot(title="Algo Total Returns")
algo_returnsbox = algo_returns.boxplot()
algo_returnsbox.plot(title="Total returns boxplot")
algo_returns.rolling(window=180).sum()
algo_returns.plot(title= "Algo Returns Rolling Returns")
algo_returns_standard_deviation = algo_returns.rolling(window=180).std()
algo_returns_standard_deviation.plot(title ="Algo Returns Standard Deviation")
algo_returnsewa = algo_returns.ewm(span=4, adjust=False).mean()
algo_returnsewa.plot(title="exponentially weighted movinng averages")
algo_returns.to_csv("algo_returnscsv")

algo_returns=algo_returns.sort_index()
```

```python
algo_returns=algo_returns.dropna().copy()
```

```python
#all the operations on the sp500_history and plotting the graphs 


sp500_path = Path('sp500_history.csv')
sp500_history=pd.read_csv(sp500_path, index_col="Date",infer_datetime_format=True,parse_dates=True)
sp500_history["Close"] = (sp500_history["Close"].str.replace("$",""))
# sp500_history.float(sp500_history.replace(',', '').replace('$', ''))
sp500_history["Close"] = sp500_history["Close"].astype("float")
sp500_history.rename(columns={ "Close" : "SP500 Price"}, inplace=True)
sp500_historypct= sp500_history.pct_change()
sp500_historypct.plot(title="SP500 Percent Change")
sp500_history.plot(title="SP500 Total Return")
sp500_historybox = sp500_history.boxplot()
sp500_historybox.plot(title="Total returns boxplot")
sp500_history.rolling(window=180).sum()
sp500_history.plot(title="SP500 Rolling Return")
sp500_history_standard_deviation = sp500_history.rolling(window=180).std()
sp500_history_standard_deviation.plot(title="SP500 Standard Deviation")
sp500_historyewm = sp500_history.ewm(span=4, adjust=False).mean()
sp500_historyewm.plot(title="exponentially weighted movinng averages")
sp500_history.to_csv("sp55_historycsv")
sp500_history = sp500_history.sort_index()
```

```python
sp500_history=sp500_history.dropna().copy()
```

```python
#all the operations on my_portfolio and plotting the graphs


my_portfoliopath = Path('MORNAndSYSCOandAAPL.csv')
my_portfolio =pd.read_csv(my_portfoliopath, index_col="Date",infer_datetime_format=True,parse_dates=True)

my_portfolio.rename(columns={'Price':'MORN'}, inplace=True)
my_portfolio.rename(columns={"Price":"SYSCO"}, inplace=True)
my_portfolio.rename(columns={"Price":"AAPL"}, inplace=True)
my_portfoliopct= my_portfolio.pct_change()
my_portfoliopct.plot(title="My Portfolio Percent Change")
my_portfolio.plot(title="My Portfolio Total Returns")
my_portfoliobox = my_portfolio.boxplot()
my_portfoliobox.plot(title="Total returns boxplot")
my_portfolio.rolling(window=180).sum()
my_portfolio.plot(title="My portfolio Rolling Returns")
my_portfolio_standard_deviation =my_portfolio.rolling(window=180).std().plot(title="my portfolio standard deviation")
my_portfolioewm = my_portfolio.ewm(span=4, adjust=False).mean()
my_portfolioewm.plot(title="exponentially weighted movinng averages")
my_portfolio.to_csv("myportfoliocsv")
my_portfolio=my_portfolio.sort_index()


print("--------------------------------------------------------------")

```

```python
#Calculatinng the Sharpe Ratio and plotting it

risk_free_rate_path = Path('risk_free_rate.csv')
risk_free_rate = pd.read_csv(risk_free_rate_path, index_col='Date', parse_dates=True, infer_datetime_format=True)
combinedlistsharpe=pd.concat([sp500_history,whale_returns, algo_returns,risk_free_rate], axis = "columns", join = "inner")
sharpe_ratiosd = ((combinedlistsharpe.mean()*combinedlistsharpe.mean()) * 252)/(combinedlistsharpe.std() * np.sqrt(252))
sharpe_ratiosd.plot.bar(figsize=(10,10))
sharpe_ratiosd.plot(title="Sharpe Ratio")
                  
```

```python
#calculat and plot the correlations 


combinedlistsharpe_correlation = combinedlistsharpe.corr()
sns.heatmap(combinedlistsharpe_correlation, vmin=-1, vmax=1)

```
