### Monte Carlo Simulation

```python
import pandas as pd
from MCForecastTools import MCSimulation as MCSimulation
# Initial imports
import os
import requests
import pandas as pd
import numpy as np
from dotenv import load_dotenv
import alpaca_trade_api as tradeapi
from MCForecastTools import MCSimulation as MCSimulation
import matplotlib.pyplot as plt
%matplotlib inline

load_dotenv()
Btc_price=0
api_version = "v2"
%matplotlib inline 
tickers = ["BTC","ETH"]
my_Btc = 1.2
my_eth=  5.3

my_agg=200
my_spy=50
monthly_income=12000
num_sims = 0
timeframe="30"
# Set start and end dates of thirty years back from today.
# Sample results may vary from the solution based on the time frame chosen
```

```python
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")

api = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version = "v2")
```

```python
start_date = pd.Timestamp('1990-08-07', tz='America/New_York').isoformat()
end_date = pd.Timestamp('2020-08-07', tz='America/New_York').isoformat()
#MCSimulation
```

```python
# Set the ticker information
MonteCarlo_tickers = ["SPY","AGG"]

timeframe= "1D"

df_portfolio = api.get_barset(
    MonteCarlo_tickers,
    timeframe,
    start=start_date,
    end=end_date
).df

# Display sample data
#df_portfolio.head()
#MC_dist.calc_cumulative_return()
```

```python
# Printing the simulation input data
# Configure a Monte Carlo simulation to forecast five years cumulative returns

MC_dist = MCSimulation(
    portfolio_data= df_portfolio,
    weights = [.50,.50],
    num_simulation = 1000,
    num_trading_days = 252)


MC_dist.calc_cumulative_return()
```

```python
# Running a Monte Carlo simulation to forecast 30 years cumulative returns
```

### Retirement Analysis

```python
# Fetch summary statistics from the Monte Carlo simulation re
MC_dist.plot_distribution()
```

### Calculate the expected portfolio return at the 95% lower and upper confidence intervals based on a `$20,000` initial investment.

```python
MC_table=MC_dist.summarize_cumulative_return()


# Set initial investment
initial_investment = 20000 * 1.5

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $30,000
# YOUR CODE HERE!

MC_ci_lower = round(MC_table[8]*15000,2)
MC_ci_upper = round(MC_table[9]*15000,2)


print(f"There is a 95% chance that an initial investment of ${initial_investment} in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${MC_ci_lower} and ${MC_ci_upper}")
```
